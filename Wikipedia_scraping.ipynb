```python
def scrape_movies_content():
  title=[]
  summary=[]
  all_movies=scrape_movie_list(2020)
  for i in all_movies:
    print(i)
    response = requests.get(f"{root}/{i}")
    content=response.text
    soup=BeautifulSoup(content,'html.parser')
    tables = soup.findAll('h1',id='firstHeading')
    title = re.findall(r'<i>(.*?)</i>', str(tables))
    summary_text = ''

    for e in soup.findAll('h2'):
      k=e.text
      if k.startswith('Plot'):
          for s in e.find_next_siblings():
            if s.name == 'h2':
              break
            if s.name == 'p':
              plot = s.get_text()
              summary_text += plot + '\n'

      summary.append(summary_text)


      #print(title)
      print(summary)



```
